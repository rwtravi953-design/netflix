```python
import pandas as pd

df=pd.read_csv(r"C:\Users\Arun\Downloads\googleplaystore.csv\googleplaystore.csv")
df["Rating"].fillna(df["Rating"].mean(),inplace=True)
df=df.drop("Current Ver",axis=1).drop("Price",axis=1)
df.dropna(inplace=True)
print(df.isnull().value_counts())
df
```

```python
from ydata_profiling import ProfileReport
report=ProfileReport(df,title="google",explorative= True)
report.to_notebook_iframe()
```

```python
x=df.drop("Rating",axis=1)
y=df["Rating"]
print(x.columns)
```

```python

from sklearn.preprocessing import PowerTransformer,OneHotEncoder,RobustScaler
import numpy as np
from category_encoders import CountEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from catboost import CatBoostRegressor


# y=np.log1p(y)


transformer=make_column_transformer(
    (OneHotEncoder(handle_unknown="ignore",sparse_output=False),['App', 'Category','Size', 'Installs', 'Genres', 'Last Updated', 'Android Ver',"Content Rating","Type"]),
    (PowerTransformer(),["Reviews"]),    
    remainder="passthrough"
)

pipeline=make_pipeline(
    
        transformer,
        CatBoostRegressor(task_type="GPU",devices="0",learning_rate=0.05)
    
)


```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)

pipeline.fit(x_train,y_train)
pred=pipeline.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import cross_val_score

print(mean_squared_error(y_test,pred))
print(r2_score(y_test,pred))

model_inside = pipeline.named_steps['catboostregressor']

print(model_inside.get_feature_importance(prettified=True))
# print(cross_val_score(pipeline,x_train,y_train,scoring='neg_mean_squared_log_error',n_jobs=-1,cv=5))
```
